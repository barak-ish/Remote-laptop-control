```python
#speech reg and text
import os
import pyttsx3
import webbrowser
import pyautogui
import speech_recognition as sr

# Function to speak text
def speak(text):
    engine = pyttsx3.init()
    engine.say(text)
    engine.runAndWait()

# Function to open a folder
def open_folder(folder_name):
    common_folders = ["downloads", "documents", "desktop"]

    folder_name_lower = folder_name.lower()

    for folder in common_folders:
        if folder_name_lower in folder:
            folder_path = os.path.join(os.path.expanduser("~"), folder)
            os.startfile(folder_path)
            print(f"Opened folder: {folder_path}")
            return

    print(f"Folder '{folder_name}' not found.")

def increase_volume():
    pyautogui.press('volumeup')

def decrease_volume():
    pyautogui.press('volumedown')

def prev_track():
    pyautogui.press('prevtrack')

def next_track():
    pyautogui.press('nexttrack')

def play_music():
    os.system("spotify")

def search_google(query):
    search_url = f"https://www.google.com/search?q={query}"
    webbrowser.open(search_url)
    print(f"Searching Google for: {query}")

# Function to recognize speech
def recognize_speech():
    recognizer = sr.Recognizer()
    with sr.Microphone() as source:
        print("Listening for a command...")
        recognizer.adjust_for_ambient_noise(source)
        audio = recognizer.listen(source)

    try:
        user_input = recognizer.recognize_google(audio).lower()
        print(f"Recognized (voice): {user_input}")
        return user_input
    except sr.UnknownValueError:
        print("Voice input not recognized.")
        return ""
    except sr.RequestError as e:
        print(f"Could not request results from Google Speech Recognition service; {e}")
        return ""

# Start a loop for continuous user input
while True:
    try:
        input_method = input("Enter 'voice' for voice input, 'text' for text input, or 'exit' to quit: ").lower()

        if input_method == "exit":
            print("Exiting program.")
            break

        elif input_method == "voice":
            user_input = recognize_speech()

        elif input_method == "text":
            user_input = input("Enter your command: ").lower()

        else:
            print("Invalid input method. Please enter 'voice', 'text', or 'exit'.")
            continue

        if user_input:
            if "open" in user_input:
                folder_name = user_input.replace("open ", "").strip()
                open_folder(folder_name)

            elif "increase volume" in user_input:
                increase_volume()

            elif "decrease volume" in user_input:
                decrease_volume()

            elif "search" in user_input:
                query = user_input.replace("search", "").strip()
                search_google(query)

            elif "play" in user_input and "music" in user_input:
                play_music()

            elif "previous" in user_input and "track" in user_input:
                prev_track()

            elif "next" in user_input:
                next_track()

    except KeyboardInterrupt:
        # Break the loop
        break
```
