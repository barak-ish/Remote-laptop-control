```python
#speech reg and text
import os
import time
import threading
import pyttsx3
import time
import webbrowser
import pyautogui
import speech_recognition as sr

# Function to speak text
def speak(text):
    engine = pyttsx3.init()
    engine.say(text)
    engine.runAndWait()

def open_folder(folder_name):
    folder_name_lower = folder_name.lower()
    folder_path = os.path.join(os.path.expanduser("~"), folder_name)
    try:
        os.startfile(folder_path)
        print(f"Opened folder: {folder_path}")
    except FileNotFoundError as e:
        print(f"Error: Folder '{folder_name}' not found.")
    except OSError as e:
        print(f"Error: An error occurred while opening the folder: {e}")

def increase_volume():
    print("Increasing volume. Type 'stop' or say 'stop' to halt.")
    stop_increasing = False

    def listen_for_stop():
        nonlocal stop_increasing
        while not stop_increasing:
            user_input = recognize_speech()  # Listen for voice input
            if user_input and "stop" in user_input.lower():
                stop_increasing = True
                print("Stopping volume increase")

    # Start a separate thread to listen for voice input
    input_thread = threading.Thread(target=listen_for_stop)
    input_thread.start()

    # Increase volume until stop is triggered
    while not stop_increasing:
        pyautogui.press('volumeup')
        time.sleep(1)

    # Wait for the input thread to finish
    input_thread.join()

def decrease_volume():
    print("Decreasing volume. Type 'stop' or say 'stop' to halt.")
    stop_increasing = False

    def listen_for_stop():
        nonlocal stop_increasing
        while not stop_increasing:
            user_input = recognize_speech()  # Listen for voice input
            if user_input and "stop" in user_input.lower():
                stop_increasing = True
                print("Stopping volume decrease")

    # Start a separate thread to listen for voice input
    input_thread = threading.Thread(target=listen_for_stop)
    input_thread.start()

    # Increase volume until stop is triggered
    while not stop_increasing:
        pyautogui.press('volumeup')
        time.sleep(1)

    # Wait for the input thread to finish
    input_thread.join()

def play_pause():
    pyautogui.press('playpause')
    print("Toggled play/pause music")

def prev_track():
    pyautogui.press('prevtrack')

def next_track():
    pyautogui.press('nexttrack')

def play_music():
    os.system("spotify")
    print("Opening Spotify to play music")


def search_google(query):
    search_url = f"https://www.google.com/search?q={query}"
    webbrowser.open(search_url)
    print(f"Searching Google for: {query}")

# Function to recognize speech
def recognize_speech():
    recognizer = sr.Recognizer()
    with sr.Microphone() as source:
        print("Listening for a command...")
        recognizer.adjust_for_ambient_noise(source)
        audio = recognizer.listen(source)

    try:
        user_input = recognizer.recognize_google(audio).lower()
        print(f"Recognized (voice): {user_input}")
        return user_input
    except sr.UnknownValueError:
        print("Voice input not recognized.")
        return ""
    except sr.RequestError as e:
        print(f"Could not request results from Google Speech Recognition service; {e}")
        return ""

def logic():
    if user_input:
        if "open" in user_input:
            folder_name = user_input.replace("open ", "").strip()
            open_folder(folder_name)

        elif "increase volume" in user_input:
            increase_volume()

        elif "decrease volume" in user_input:
            decrease_volume()

        elif "search" in user_input:
            query = user_input.replace("search", "").strip()
            search_google(query)

        elif "pause" in user_input or "continue" in user_input:
            play_pause()

        elif "play" in user_input and "music" in user_input:
            play_music()

        elif "previous" in user_input and "track" in user_input:
            prev_track()

        elif "next" in user_input:
            next_track()
        else:
            print("Command not recognized. Please try again.")
# Start a loop for continuous user input

while True:
    try:
        input_method = input("Enter 'voice' for voice input, 'text' for text input, or 'exit' to quit: ").lower()

        if input_method == "exit":
            print("Exiting program.")
            break

        elif input_method == "voice":
            while True:
                user_input = recognize_speech()
                logic()

                if user_input.lower() == "exit":
                    print("Exiting voice input mode.")
                    break

        elif input_method == "text":
            while True:
                user_input = input("Enter your command: ").lower()
                logic()
                if user_input == "exit":
                    print("Exiting text input mode.")
                    break
        else:
            print("Invalid input method. Please enter 'voice', 'text', or 'exit'.")

    except KeyboardInterrupt:
        # Break the loop if the user presses Ctrl+C
        break
```
